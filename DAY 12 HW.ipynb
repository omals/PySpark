```python
appointments = [
    (101, " arjun ", "cardiology", "2024-01-05", 5000),
    (102, "MEERA", "neurology", "2024-01-12", 7000),
    (103, "rahul", "orthopedics", "2024-02-03", 4500),
    (104, " Sneha", "dermatology", "2024-02-18", 3000),
    (105, "vikram ", "cardiology", "2024-03-01", 8000),
    (106, "ASHA", "neurology", "2024-03-10", 6500),
    (107, " kiran ", "orthopedics", "2024-03-22", 5000),
    (108, "Neha", "dermatology", "2024-04-05", 3500),
    (109, "john ", "cardiology", "2024-04-15", 7200),
    (110, " PRIYA", "neurology", "2024-04-25", 6800)
]
 
columns = [
    "appointment_id",
    "patient_name",
    "department",
    "appointment_date",
    "bill_amount"
]
 
df_appointments = spark.createDataFrame(appointments, columns)
 
df_appointments.display()
```

# Level 1: Date Functions
## Task 1
- Convert: appointment_date → proper date type


```python
from pyspark.sql.functions import to_date
df_appointments.withColumn("appointment_date",to_date("appointment_date")).display()
```

##  Task 2
 Extract:
year from appointment_date


```python
from pyspark.sql.functions import to_date, year
df_appointments.withColumn("year",year(to_date("appointment_date"))).display()
```

##  Task 3
 Extract:
month from appointment_date


```python
from pyspark.sql.functions import month, to_date
df_appointments.withColumn("month",month(to_date("appointment_date"))).display()
```

##  Task 4
 Extract:
day from appointment_date


```python
from pyspark.sql.functions import to_date, dayofmonth
df_appointments.withColumn("day",dayofmonth(to_date("appointment_date"))).display()
```

# Level 2: String Functions
##  Task 5
 Convert:
patient_name → uppercase


```python
from pyspark.sql.functions import upper
df_appointments.withColumn("patient_name", upper("patient_name")).display()
```

##  Task 6
 Convert:
department → uppercase


```python
from pyspark.sql.functions import upper
df_appointments.withColumn("department", upper("department")).display()
```

##  Task 7
 Remove extra spaces from patient_name


```python
from pyspark.sql.functions import trim
df_appointments.withColumn("patient_name", trim("patient_name")).display()
```

##  Task 8
 Create:
patient_department

(example: ARJUN - CARDIOLOGY)


```python
from pyspark.sql.functions import concat, lit
df_appointments.withColumn("patient_department",concat(upper("patient_name"),lit(" - "),upper("department"))).display()
```

# Level 3: Analysis Tasks
##  Task 9
 Find:
total billing per department


```python
from pyspark.sql.functions import sum
df_appointments.groupBy("department").agg(sum("bill_amount").alias("total_bill")).display()
```

##  Task 10
 Find:
total billing per month

 
 

```python
df_appointments.groupBy(month(to_date("appointment_date")).alias("month")).agg(sum("bill_amount").alias("total_bill")).display()
```

##  Task 11
 Find:
department with highest billing

```python
from pyspark.sql.functions import desc, sum
df_appointments.groupBy("department").agg(sum("bill_amount").alias("total_bill")).orderBy(desc("total_bill")).first()
```

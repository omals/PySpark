```python
df = spark.read.json("/Volumes/workspace/new/patient_json/ptient.json")
df.show()
```

```python
df.printSchema()
```

```python
from pyspark.sql.functions import *

df = df.withColumn("bill_amount", col("bill_amount").cast("int"))
df = df.withColumn("patient_id", col("patient_id").cast("int"))
df.printSchema()
```

```python
df = df.filter(isnull("bill_amount"))
df.show()
```

```python
df.dropDuplicates()
```

```python
df.write.mode("overwrite").saveAsTable("cleaned_patient")
```

```python
customers = [
    (1, "Arjun", "Kochi"),
    (2, "Meera", "Chennai"),
    (3, "Rahul", "Bangalore"),
    (4, "Sneha", "Kochi")
]
 
columns = ["customer_id", "name", "city"]
 
df_customers = spark.createDataFrame(customers, columns)
df_customers.show()
```

```python
orders = [
    (101, 1, 5000),
    (102, 2, 3000),
    (103, 1, 2000),
    (104, 5, 4000)
]
 
columns = ["order_id", "customer_id", "amount"]
 
df_orders = spark.createDataFrame(orders, columns)
df_orders.show()
```

```python
df_customers.join(df_orders,"customer_id","inner").show()
```

```python
df_customers.join(df_orders,"customer_id","left").show()
```

```python
df_customers.join(df_orders,"customer_id","right").show()
```

```python
df_orders.join(df_customers,"customer_id","left").groupBy("city").sum("amount").show()
```

```python
df_orders.join(df_customers,"customer_id","inner").groupBy("city").sum("amount").show()
```

```python
df = df_orders.join(df_customers,"customer_id","left")
df.filter(df.city != "NULL").groupBy("city").sum("amount").show()
```

```python
df_patient = spark.table("workspace.new.data_2")
df_patient.display()
```

- Q1 Create: column "age_category" <30 → Young, 30–50 → Adult and 50 → Senior

```python

from pyspark.sql.functions import when

df_patient.withColumn("age_category",when(df_patient.age < 30,"Young").when((df_patient.age >= 30) & (df_patient.age <= 50) ,"Adult").when(df_patient.age >50, "Senior")).show()
```

- Q2 Create: column "city_flag" Kochi → "Local", Others → "Non-Local"

```python
from pyspark.sql.functions import when

df_patient.withColumn("city_flag",when(df_patient.city == "Kochi", "Local").otherwise("Non-Local")).show()
```

- Q3 Create: "special_attention" Emergency + age > 50 → Yes Else → No

```python
from pyspark.sql.functions import when

df_patient.withColumn("special_attention", when((df_patient.admission_type == "Emergency") & (df_patient.age > 50), "Yes").otherwise("No")).show()
```

- Q4 Create column: city_type Kochi → "Metro" Others → "Non-Metro"

```python
df_patient.withColumn("city_type",when(df_patient.city == "Kochi","Metro").otherwise("Non-Metro")).show()
```

- Q5  Add column: is_test_data = False

```python
from pyspark.sql.functions import lit

df_patient.withColumn("is_test_data", lit("False")).show()
```

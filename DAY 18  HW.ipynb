#  Level 4: CASE WHEN
## Task 12
Create:
sales_category
Rules:

< 20000 → Low
20000 - 50000 → Medium
> 50000 → High

```python
spark.sql(""" SELECT order_id, product, category, city, sales_amount, CASE WHEN sales_amount < 20000 THEN 'Low' WHEN sales_amount between 20000 AND 50000 THEN 'Medium' ELSE 'High' END AS sales_category FROM orders """).show()
```

## Task 13
Count:
How many records belong to each sales_category.

```python
spark.sql(""" SELECT sales_category, COUNT(*) as category_count FROM ( SELECT CASE WHEN sales_amount < 20000 THEN 'Low' WHEN sales_amount between 20000 AND 50000 THEN 'Medium' ELSE 'High' END AS sales_category FROM orders) GROUP BY sales_category """).show()
```

#  Level 5: Window Functions in SQL
## Task 14
Rank products within each city based on sales_amount.

```python
spark.sql(""" SELECT city, product, RANK() OVER (PARTITION BY city ORDER BY sales_amount DESC) AS rank FROM orders """).show()
```

```python
spark.sql(""" WITH product_sales AS ( SELECT city, product, SUM(sales_amount) AS total_sales FROM orders GROUP BY city, product)   SELECT city, product, total_sales, RANK() OVER ( PARTITION BY city ORDER BY total_sales DESC ) AS rank FROM product_sales """).show()
```

## Task 15
Find:
Top product in each city.

```python
spark.sql("""
WITH product_sales AS ( SELECT city, product, SUM(sales_amount) AS total_sales FROM orders GROUP BY city, product ),
ranked_products AS ( SELECT city, product, total_sales, RANK() OVER ( PARTITION BY city ORDER BY total_sales DESC ) AS rank FROM product_sales )
SELECT city, product, total_sales, rank FROM ranked_products WHERE rank = 1 """).show()
```

## Task 16

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, rand, round

spark = SparkSession.builder.getOrCreate()

df = spark.range(1, 21).withColumn("age", (rand()*40 + 20).cast("int")) \
                      .withColumn("salary", round(rand()*50000 + 30000, 2)) \
                      .withColumn("department", 
                                  (col("id") % 3).cast("int"))

df.show()
```

Select only id and salary from the DataFrame.

```python
df.select("id", "salary").show()
```

Add a new column bonus = 10% of salary using withColumn().

```python
df.withColumn("bonus", col("salary") * 0.10).show()
```

Create a new DataFrame with only id and a calculated column salary_after_tax (salary * 0.9) using select().

```python
df_salary_after_tax = df.select("id", (col("salary") * 0.9).alias("salary_after_tax"))
df_salary_after_tax.show()
```

While selecting, rename department as dept_id.

```python
df.select("id", "salary", col("department").alias("dept_id")).show()
```

Use alias inside an aggregation (e.g., average salary as avg_sal).

```python
from pyspark.sql.functions import avg

df.select(avg(col("salary")).alias("avg_sal")).show()
```

Find unique combinations of age and department.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, rand, round

spark = SparkSession.builder.getOrCreate()

df = spark.range(1, 21).withColumn("age", (rand()*40 + 20).cast("int")) \
                      .withColumn("salary", round(rand()*50000 + 30000, 2)) \
                      .withColumn("department", 
                                  (col("id") % 3).cast("int"))

df.display()

df.dropDuplicates(["age", "department"]).select("id","age", "department").display()
```

Use both orderBy() and sort() for the same operation and compare outputs.

```python
df.orderBy("age", "department").show()
df.sort("age", "department").show()
```

Combine orderBy() and limit() to get the youngest 5 employees.

```python
df.orderBy("age").limit(5).show()
```

Count how many people have salary greater than 50,000.

```python
df.filter(col("salary") > 50000).count()
```

Count distinct age values.

```python
df.select("age").distinct().count()
```

```python
data = [
    (1, " arjun ", "cardiology", "2024-01-01", 5000),
    (2, "MEERA", "neurology", "2024-01-03", 7000),
    (3, "rahul", "orthopedics", "2024-02-10", 4000),
    (4, "Sneha", "dermatology", "2024-02-15", 3500),
    (5, "vikram ", "cardiology", "2024-03-01", 8000)
]
 
columns = ["patient_id", "patient_name", "department", "visit_date", "bill_amount"]
 
df = spark.createDataFrame(data, columns)
 
df.display()
```

```python
from pyspark.sql.functions import to_date
df.withColumn("visit_date", to_date("visit_date")).display()
```

```python
from pyspark.sql.functions import year
df.withColumn("year", year("visit_date")).display()
```

```python
from pyspark.sql.functions import month
df.withColumn("month",month("visit_date")).display()
```

```python
from pyspark.sql.functions import dayofmonth
df.withColumn("day",dayofmonth("visit_date")).display()
```

```python
from pyspark.sql.functions import upper
df.withColumn("department", upper("department")).display()
```

```python
from pyspark.sql.functions import lower
df.withColumn("patient_name",lower("patient_name")).display()
```

```python
from pyspark.sql.functions import trim
df.withColumn("clean_name",trim("patient_name")).display()
```

```python
from pyspark.sql.functions import concat
df.withColumn("patient_info", concat("patient_name","department")).display()
```

```python
from pyspark.sql.functions import concat,lit
df.withColumn("patient_info", concat("patient_name",lit("-"),"department")).display()
```

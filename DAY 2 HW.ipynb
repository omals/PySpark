Basic Filters
 - Q1. Show patients with: age > 40
 - Q2. Show all Female patients
 - Q3. Show patients from Kochi

```python
df_patient = spark.table("workspace.new.data_2")
df_patient.show()
df_patient.filter(df_patient.age > 40).show()
df_patient.filter(df_patient.gender == 'Female').show()
df_patient.filter(df_patient.city == 'Kochi').display()
```

Intermediate Filters

- Q1 Show patients: age > 35 AND city = Kochi
- Q2 Show patients: department = Cardiology
- Q3 Show patients: admission_type = Emergency

```python
df_patient.filter((df_patient.age > 35) & (df_patient.city == "Kochi")).display()
df_patient.filter(df_patient.department == "Cardiology").display()
df_patient.filter(df_patient.admission_type == "Emergency").display()
```

Combined Select + Filter

- Q1 Show name and city where age > 40
- Q2 Show name, department where gender = Female
- Q3 Show name, age where city = Thrissur

```python
df_patient.filter(df_patient.age > 40).select("name","city").display()
df_patient.filter(df_patient.gender == "Female").select("name","department").display()
df_patient.filter(df_patient.city == "Thrissur").select("name", "age").display()
```

Logical Conditions 

- Q1 Show patients Cardiology AND Kochi
- Q2 Show patients Neurology OR Orthopedics
- Q3 Show patients age > 30 AND gender = Male

```python
df_patient.filter((df_patient.department == "Cardiology") & (df_patient.city == "Kochi")).display()
df_patient.filter((df_patient.department == "Neurology") | (df_patient.department == "Orthopedics")).display()
df_patient.filter((df_patient.age > 30) & (df_patient.gender == "Male")).display()
```

Real Thinking Questions 

- Q1 Show name, department patients from Kochi with Emergency admission
- Q2 Show Female patients older than 30
- Q3 Show name, city patients NOT from Kochi

```python
df_patient.filter((df_patient.city == "Kochi") & (df_patient.admission_type == "Emergency")).select("name", "department").display()
df_patient.filter((df_patient.gender == "Female") & (df_patient.age > 30)).display()
df_patient.filter(df_patient.city != "Kochi").select("name", "city").display()
```

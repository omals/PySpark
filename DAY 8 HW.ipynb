```python
sales_data = [
    (1, "Kochi", "Laptop", 50000),
    (2, "Kochi", "Mobile", 30000),
    (3, "Kochi", "Tablet", 30000),
    (4, "Kochi", "TV", 60000),
    (5, "Kochi", "Headphones", 10000),
 
    (6, "Chennai", "Laptop", 45000),
    (7, "Chennai", "Mobile", 45000),
    (8, "Chennai", "Tablet", 20000),
    (9, "Chennai", "TV", 70000),
    (10, "Chennai", "Headphones", 15000),
 
    (11, "Bangalore", "Laptop", 60000),
    (12, "Bangalore", "Mobile", 60000),
    (13, "Bangalore", "Tablet", 35000),
    (14, "Bangalore", "TV", 80000),
    (15, "Bangalore", "Headphones", 20000),
 
    (16, "Hyderabad", "Laptop", 55000),
    (17, "Hyderabad", "Mobile", 25000),
    (18, "Hyderabad", "Tablet", 25000),
    (19, "Hyderabad", "TV", 75000),
    (20, "Hyderabad", "Headphones", 12000),
 
    (21, "Delhi", "Laptop", 65000),
    (22, "Delhi", "Mobile", 40000),
    (23, "Delhi", "Tablet", 40000),
    (24, "Delhi", "TV", 85000),
    (25, "Delhi", "Headphones", 18000)
]
 
columns = ["order_id", "city", "product", "revenue"]
 
df_sales_hw = spark.createDataFrame(sales_data, columns)
df_sales_hw.show()
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, rank, desc, dense_rank, count
```

# Core Ranking
- Q1 Rank all products globally using row_number()
 

```python
df_sales_hw.withColumn("rank", row_number().over(Window.orderBy(desc("revenue" )))).show()
```

- Q2 Rank all products globally using rank()

```python
df_sales_hw.withColumn("rank", rank().over(Window.orderBy(desc("revenue" )))).show()
```

- Q3 Rank all products globally using dense_rank()

```python
df_sales_hw.withColumn("rank", dense_rank().over(Window.orderBy(desc("revenue" )))).show()
```

# Partition Ranking
- Q4 Rank products within each city using row_number()


```python
df_sales_hw.withColumn("rank", row_number().over(Window.partitionBy("city").orderBy(desc("revenue")))).show()
```

- Q5 Rank products within each city using rank()


```python
df_sales_hw.withColumn("rank", rank().over(Window.partitionBy("city").orderBy(desc("revenue")))).show()
```

- Q6 Rank products within each city using dense_rank()

```python
df_sales_hw.withColumn("rank", dense_rank().over(Window.partitionBy("city").orderBy(desc("revenue")))).show()
```

# Top-N Problems
- Q7 Find: Top product in each city


```python
df_sales_hw.withColumn("rank", row_number().over(Window.partitionBy("city").orderBy(desc("revenue")))).filter("rank==1").show()
```

- Q8 Find: Top 2 products per city


```python
df_sales_hw.withColumn("rank", row_number().over(Window.partitionBy("city").orderBy(desc("revenue")))).filter("rank<=2").show()
```

- Q9 Find: Top 3 products per city

```python
df_sales_hw.withColumn("rank", row_number().over(Window.partitionBy("city").orderBy(desc("revenue")))).filter("rank<=3").show()
```

# Tie Handling 
- Q10 Find: All top products (including ties) in each city


```python
df_sales_hw.withColumn("rank", rank().over(Window.partitionBy("city").orderBy(desc("revenue")))).filter("rank == 1").show()
```

- Q11 Find: Cities where multiple products share rank 1

```python
df_sales_hw.withColumn("rank", rank().over(Window.partitionBy("city").orderBy(desc("revenue")))).filter("rank == 2").groupBy("city").agg(count("*").alias("count")).filter("count >1").show()
```

# Slightly Tough
- Q12 Find: Second highest product in each city


```python
df_sales_hw.withColumn("rank", dense_rank().over(Window.partitionBy("city").orderBy(desc("revenue")))).filter("rank == 2").show()
```

- Q13 Find: Products ranked between 2 and 4 in each city

```python
df_sales_hw.withColumn("rank", dense_rank().over(Window.partitionBy("city").orderBy(desc("revenue")))).filter("rank BETWEEN 2 AND 4").show()
```

# Advanced Thinking
- Q14 Find: Top product overall across all cities


```python
df_sales_hw.orderBy(desc("revenue")).select("product","revenue").show(1)
```

- Q15 Find: Top city based on highest product revenue

```python
from pyspark.sql.functions import max
df_sales_hw.groupBy("city").agg(max("revenue").alias("max_revenue")).orderBy(desc("max_revenue")).show(1)
```

# Challenge 
- Q16 Find:
    For each city: product contributing highest revenue difference from second product

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, desc, when, col, max

df = df_sales_hw.withColumn("rank",row_number().over(Window.partitionBy("city").orderBy(desc("revenue")))).filter("rank <= 2")

df_result = df.groupBy("city").agg(
    max(when(col("rank") == 1, col("product"))).alias("top_product"),
    max(when(col("rank") == 1, col("revenue"))).alias("top_revenue"),
    max(when(col("rank") == 2, col("revenue"))).alias("second_revenue"))

df_final = df_result.withColumn("revenue_diff",col("top_revenue") - col("second_revenue")).orderBy(desc("revenue_diff")).show()
```

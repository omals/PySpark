```python
df_stud = spark.table("workspace.new.data_1")

df_stud.show()

df_stud.printSchema()

display(df_stud.limit(10))
display(df_stud.head(10))
print("There are ",len(df_stud.columns)," columns in the table.")

print("The data type of course is ",df_stud.schema["age"].dataType)

course_counts = df_stud.groupBy("course").count().orderBy("count", ascending=False)
print("Course appears the most is ",course_counts.first()[0])
```

load data as csv , write the function used and its components

```python
df_patient = spark.read.csv(
    "/Volumes/workspace/new/csv_files/Data1.csv",
    header=True,
    inferSchema=True,
    sep=','
)

df_patient.show()
df_patient.head(2)
```

```python
from pyspark.sql.functions import sum as _sum

print(df_patient.count())
print(df_patient.columns)
print(df_patient.schema["gender"].dataType)

df_pat = df_patient.withColumn("age", df_patient["age"].cast("integer"))
total_age = df_pat.agg(_sum("age").alias("sum(total"))
# total_age.show()
print(total_age.head())
```

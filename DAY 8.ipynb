```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number, rank,desc
```

```python
sales = [
    (1, "Kochi", "Laptop", 50000),
    (2, "Kochi", "Mobile", 20000),
    (3, "Kochi", "Tablet", 30000),
    (4, "Chennai", "Laptop", 60000),
    (5, "Chennai", "Mobile", 25000),
    (6, "Chennai", "Tablet", 20000),
    (7, "Bangalore", "Laptop", 60000),
    (8, "Bangalore", "Mobile", 15000),
    (9, "Bangalore", "Tablet", 35000)
]
 
columns = ["order_id", "city", "product", "revenue"]
 
df_sales = spark.createDataFrame(sales, columns)
df_sales.show()
```

```python
df_sales.withColumn("rank",row_number().over(Window.partitionBy("city").orderBy(desc("revenue")))).show()
```

```python
df_sales.withColumn("rank",rank().over(Window.partitionBy("product").orderBy(desc("revenue")))).show()
```

```python
from pyspark.sql.functions import dense_rank
df_sales.withColumn("rank",dense_rank().over(Window.partitionBy("product").orderBy(desc("revenue")))).show()
```

```python
data = [

    (101, "Laptop", "Kochi", 50000),

    (102, "Mobile", "Kochi", 20000),

    (103, "Tablet", "Chennai", 30000),

    (104, "Laptop", "Chennai", 45000),

    (105, "Mobile", "Bangalore", 15000),

    (106, "Laptop", "Bangalore", 60000),

    (107, "Tablet", "Hyderabad", 25000),

    (108, "Laptop", "Hyderabad", 55000),

    (109, "TV", "Delhi", 80000),

    (110, "Mobile", "Delhi", 40000)

]
 
columns = [

    "order_id",

    "product",

    "city",

    "sales_amount"

]
 
df = spark.createDataFrame(data, columns)
 
```

```python
df.show()
```

```python
df.createOrReplaceTempView("orders")
```

```python
spark.sql(""" SELECT * FROM orders """).show()
```

```python
spark.sql(""" SELECT * FROM ORDERS WHERE sales_amount > 40000""").show()
```

```python
spark.sql(""" SELECT city, SUM(sales_amount) FROM orders GROUP BY city""").show()
```

```python
spark.sql(""" SELECT * , dense_rank() OVER(PARTITION BY city ORDER BY sales_amount DESC) as rank FROM orders """).show()
```

```python
data = [
    (1, "Arjun", 50000),
    (1, "Arjun", 50000),
    (2, "Meera", 30000),
    (3, "Rahul", 40000),
    (3, "Rahul", 40000)
]
 
df = spark.createDataFrame(data, ["id","name","salary"])
df.show()
```

```python
df.dropDuplicates().show()
```

```python
data = [
    (1, "Arjun", None),
    (2, "Meera", 30000),
    (3, "Rahul", None)
]
 
df = spark.createDataFrame(data, ["id","name","salary"])
df.show()
```

```python
df.na.drop().show()
```

```python
df.fillna({"salary":0}).show()
```

```python
from pyspark.sql.functions import when
df.withColumn("salary_band", when(df.salary > 20000, "high")).show()
```

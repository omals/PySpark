```python
sales_data = [
    (101, "Kochi", "Medicine", 1200),
    (102, "Kochi", "Medicine", 1800),
    (103, "Kochi", "Supplements", 2500),
    (104, "Kochi", "Medical Devices", 4500),
    (105, "Chennai", "Medicine", 1500),
    (106, "Chennai", "Supplements", 2200),
    (107, "Chennai", "Medical Devices", 5000),
    (108, "Chennai", "Medicine", 1700),
    (109, "Bangalore", "Medicine", 2100),
    (110, "Bangalore", "Supplements", 2800),
    (111, "Bangalore", "Medical Devices", 6000),
    (112, "Bangalore", "Medicine", 2400),
    (113, "Hyderabad", "Medicine", 1600),
    (114, "Hyderabad", "Supplements", 2600),
    (115, "Hyderabad", "Medical Devices", 5500),
    (116, "Hyderabad", "Medicine", 1900),
    (117, "Delhi", "Medicine", 2300),
    (118, "Delhi", "Supplements", 3200),
    (119, "Delhi", "Medical Devices", 7000),
    (120, "Delhi", "Medicine", 2700),
    (121, "Kochi", "Supplements", 2900),
    (122, "Chennai", "Supplements", 3100),
    (123, "Bangalore", "Medicine", 2600),
    (124, "Hyderabad", "Medical Devices", 5800),
    (125, "Delhi", "Supplements", 3400)
]
 
columns = [
    "sale_id",
    "city",
    "product_category",
    "sales_amount"
]
 
df_sales = spark.createDataFrame(sales_data, columns) 
df_sales.show()
 
```

```python
df_sales.printSchema()
```

```python
from pyspark.sql.functions import *
```

#  Exercises for This Dataset
##  Task 1
- Find total sales by product category.

```python
df_sales.groupBy("product_category").agg(sum("sales_amount").alias("total")).show()
```

##  Task 2
- Find average sales amount by product category.

```python
df_sales.groupBy("product_category").agg(avg("sales_amount").alias("average")).show()
```

##  Task 3
- Find number of sales transactions by product category.

```python
df_sales.groupBy("product_category").agg(count("*").alias("count")).show()
```

##  Task 4
- Create a report showing:
| Product Category | Total Sales | Avg Sales | Transaction Count |

```python
report = df_sales.groupBy("product_category").agg(sum("sales_amount").alias("Total_sales"), avg("sales_amount").alias("Avg_sales"),count("*").alias("Transaction_count"))
report.show()
```

## Task 5
- Find total sales by city.

```python
df_sales.groupBy("city").agg(sum("sales_amount").alias("total")).show()
```

##  Task 6
Find average sales amount by city.

```python
df_sales.groupBy("city").agg(avg("sales_amount").alias("average")).show()
```

##  Task 7
- Create a pivot table:
| Product Category | Kochi | Chennai | Bangalore | Hyderabad | Delhi |
Showing total sales.

```python
pivot_table = df_sales.groupBy("product_category").pivot("city",["Kochi", "Chennai", "Bangalore", "Hyderabad", "Delhi"]).agg(sum("sales_amount"))
pivot_table.show()
```

##  Task 8
- Replace null values in the pivot table with 0.

```python
pivot_table = pivot_table.fillna(0)
pivot_table.show()
```

##  Task 9
- Find the city with the highest sales.

```python
df_sales.groupBy("city").agg(sum("sales_amount").alias("total")).orderBy(desc("total")).show(1)
```

##  Task 10
- Calculate KPI Metrics:
    - Total Sales
    - Total Transactions
    - Average Transaction Value

```python
kpi_metrics = df_sales.agg(sum("sales_amount").alias("Total Sales"), count("*").alias("Total Transactions"), avg("sales_amount").alias("Average Transaction Value"))
kpi_metrics.show()
```

#  Challenge Tasks
## Task 11
- Create a report:
| Product Category | Revenue | Avg Sales | Transactions |
Sort by Revenue descending.

```python
report_df = df_sales.groupBy("product_category").agg(sum("sales_amount").alias("Revenue"), avg("sales_amount").alias("Avg  sales"), count("*").alias("Transactions")).orderBy(desc("Revenue"))
report_df.show()
```

## Task 12
- Create a pivot:
City vs Product Category
Showing transaction count.

```python
pivot_tb = df_sales.groupBy("city").pivot("product_category").agg(count("sale_id"))
pivot_tb.show()
```

## Task 13
Which product category contributes the highest percentage of total sales?
 
 

```python
total = df_sales.agg(sum("sales_amount")).collect()[0][0]
print(total)
percent = df_sales.groupBy("product_category").agg(sum("sales_amount").alias("Revenue")).withColumn("percentage", round(( col("Revenue")/total)*100, 2))
percent.show()
percent.orderBy(desc("percentage")).show(1)
```

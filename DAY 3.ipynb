```python
df_patient = spark.table("workspace.new.data_2")
df_patient.show()
```

```python
df_patient.withColumn("age_plus_10", df_patient.age + 10).show()
```

```python
from pyspark.sql.functions import when

df_patient.withColumn("result",when(df_patient.age<30,"Young").when((df_patient.age>=30) & (df_patient.age <50),"Middle").otherwise("senior")).show()
```

```python
from pyspark.sql.functions import when

df_patient.withColumn("gender_label", when(df_patient.gender == "Male", "M").otherwise("F")).show()
```

```python
from pyspark.sql.functions import when

df_patient.withColumn("is_kochi", when(df_patient.city == "Kochi",1).otherwise("0")).show()
```

```python
from pyspark.sql.functions import when

df_patient.withColumn("is_kochi", when(df_patient.city == "Kochi",1)).show()
```

```python
from pyspark.sql.functions import lit

df_patient.withColumn("country",lit("India")).show()
```

```python
from pyspark.sql.functions import lit

df_patient.withColumn("hospital_id",lit("101")).show()
```

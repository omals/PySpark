```python
df_patient = spark.table("workspace.new.data_2")
df_patient.display()
```

```python
df_patient.groupby("department").count().show()
```

```python
from pyspark.sql.functions import avg,count
df_patient.groupBy("department").agg(avg("age")).show()
df_patient.groupBy("department").agg(avg("age"),count("*")).show()
```

```python
df_patient.groupBy("department").agg(avg("age").alias("average_age")).show()
```

```python
df_patient.groupBy("department").agg(avg("age").alias("average_age")).filter("average_age> 40").show()
```

```python
fl = df_patient.groupBy("department").agg(avg("age").alias("average_age"),count("*"))
fl.filter("average_age> 40").show()
```

```python
from pyspark.sql.functions import desc
df_patient.groupBy("department").agg(count("*").alias("count")).orderBy(desc("count")).show()
```

```python
df_patient.groupBy("city").agg(count("city").alias("city_count")).orderBy(desc("city_count")).show(1)
```

```python
df_patient.groupBy("city").agg(count("city").alias("city_count")).orderBy(desc("city_count")).tail(1)
```

```python
df_patient.groupBy("department").agg(avg("age").alias("average_age")).orderBy(desc("average_age")).show(1)
```

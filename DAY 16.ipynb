```python
data = [

    (1, "Cardiology", "Kochi", 5000),

    (2, "Cardiology", "Kochi", 7000),

    (3, "Neurology", "Chennai", 6000),

    (4, "Neurology", "Chennai", 8000),

    (5, "Orthopedics", "Bangalore", 4500),

    (6, "Orthopedics", "Bangalore", 5500),

    (7, "Cardiology", "Chennai", 7500),

    (8, "Neurology", "Kochi", 6500),

    (9, "Dermatology", "Delhi", 3500),

    (10, "Dermatology", "Delhi", 4000),

    (11, "Cardiology", "Delhi", 9000),

    (12, "Orthopedics", "Kochi", 5000)

]
 
columns = [

    "patient_id",

    "department",

    "city",

    "bill_amount"

]
 
df = spark.createDataFrame(data, columns)
 
df.show()
 
```

```python
from pyspark.sql.functions import *
```

```python
df.groupBy("department").pivot("city").agg(sum("bill_amount")).show()
```

```python
df.groupBy("department").pivot("city").agg(sum("bill_amount")).fillna(0).show()
```
